```python
from pathlib import Path
birdspath = Path("Birds")
birdspath.ls()
```

```python
import re
all_birds_Name = [re.sub(r".csv|\.\s|\s*|\’" , r"", x.name) for x in birdspath.ls()]
all_birds_Name
```

```python
folder_file = {x:y for x ,y in zip(all_birds_Name , birdspath.ls())}
folder_file
```

```python
for folder,file in folder_file.items():
    path = Path("data/birds")
    dest = path/folder
    dest.mkdir(parents=True,exist_ok=True)
    download_images(file, dest)
```

```python
birdsPerClassCount = { (path/folder).name :len((path/folder).ls()) for folder,file in folder_file.items() }
dict(sorted(birdsPerClassCount.items() , key=lambda x :x[1]))
```

```python
for folder,file in folder_file.items():
    path = Path("data/birds")
    verify_images(path/folder,delete=True,max_workers=8)
```

```python
birdsPerClassCountAfterVerify = { (path/folder).name :len((path/folder).ls()) for folder,file in folder_file.items() }
dict(sorted(birdsPerClassCountAfterVerify.items() , key=lambda x :x[1]))
```

```python
data = ImageDataBunch.from_folder(path ,train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.show_batch(rows=5, figsize=(15,15))
```

```python
data.classes
```

```python
learn = create_cnn(data, models.resnet34, metrics=accuracy)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.recorder.plot_metrics()
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(5e-5,1e-4))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9 ,figsize=(15,15))
```
